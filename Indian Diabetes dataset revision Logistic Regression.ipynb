It seems that there are no missing values in out data. Great, let's see the distribution of data.

min is zero in Glucose, BloodPressure, SkinThickness, Insulin, BMI which makes no sense since these columns cannot be zero for any human being. This means something is wrong with the data.

higher variance is observed in Insulin column where mean is 79 and std is 115.

```python
# Let's see how data is distributed for every column
plt.figure(figsize=(20,25), facecolor='white')
plotnumber=1

for column in data:
    if plotnumber<=9:   # as there are 9 columns in data
        ax=plt.subplot(3,3,plotnumber)
        sns.distplot(data[column])
        plt.xlabel(column,fontsize=20)
        #plt.ylabel('Salary',fontsize=20)
    plotnumber+=1
plt.show()    
```

```python
# replacing zero values with the mean of the column
data['BMI']=data['BMI'].replace(0,data['BMI'].mean())
data['BloodPressure']=data['BloodPressure'].replace(0,data['BloodPressure'].mean())
data['Glucose']=data['Glucose'].replace(0,data['Glucose'].mean())
data['Insulin']=data['Insulin'].replace(0,data['Insulin'].mean())
data['SkinThickness']=data['SkinThickness'].replace(0,data['SkinThickness'].mean())
```

```python
# Let's see how data is distributed for every column
plt.figure(figsize=(20,25), facecolor='white')
plotnumber=1

for column in data:
    if plotnumber<=9:   
        ax=plt.subplot(3,3,plotnumber)
        sns.distplot(data[column])
        plt.xlabel(column,fontsize=20)
        #plt.ylabel('Salary',fontsize=20)
    plotnumber+=1
plt.show()    
```

Now we have dealt with the 0 values and data looks better. But, there still some outliers are present in some columns. Let's deal with them.

```python
fig, ax=plt.subplots(figsize=(15,10))
sns.boxplot(data=data, width=0.5, ax=ax, fliersize=3)
```

```python
q= data['Pregnancies'].quantile(0.98)
# we are removing the top 2% data from the Pregnancies column
data_cleaned= data[data['Pregnancies']<q]
q= data_cleaned['BMI'].quantile(0.99)
# we are removing the top 1% data from the BMI column
data_cleaned= data_cleaned[data_cleaned['BMI']<q]
q=data_cleaned['SkinThickness'].quantile(0.99)
# we are removing the top 1% data from the SkinThickness column.
data_cleaned=data_cleaned[data_cleaned['SkinThickness']<q]
q= data_cleaned['Insulin'].quantile(0.95)
# we are removing the top 5% data from the insulin column
data_cleaned=data_cleaned[data_cleaned['Insulin']<q]
q= data_cleaned['DiabetesPedigreeFunction'].quantile(0.99)
# we are removing the top 1% data from the DiabetesPedigreeFunciton
data_cleaned= data_cleaned[data_cleaned['DiabetesPedigreeFunction']<q]
q=data_cleaned['Age'].quantile(0.99)
# we are removing the top 1% data from the Age Column
data_cleaned=data_cleaned[data_cleaned['Age']<q]
```

```python
# Let's see how data is distributed for every column
plt.figure(figsize=(20,25), facecolor='white')
plotnumber=1

for column in data:
    if plotnumber<=9:   
        ax=plt.subplot(3,3,plotnumber)
        sns.distplot(data[column])
        plt.xlabel(column,fontsize=20)
        #plt.ylabel('Salary',fontsize=20)
    plotnumber+=1
plt.show()    
```

The data looks much better now than before. we will start our analysis with this data now as we dont want to lose important info. If our model doesnt work with accuracy, we will come back for more processing.

```python
X= data.drop(columns=['Outcome'])
y=data['Outcome']
```

Before we fit our data to a model, let's visualize the relationship between our independent variables and the categories.

```python
# Let's see how data is distributed for every column
plt.figure(figsize=(20,25), facecolor='white')
plotnumber=1

for column in X:
    if plotnumber<=9:   
        ax=plt.subplot(3,3,plotnumber)
        sns.stripplot(y,X[column])
    plotnumber+=1
plt.tight_layout()    
```

Great!! Let's proceed by checking multicollinearity in the dependent variables. Before that, we should scale our data. Let's use the std. scaler for that.

```python
scalar= StandardScaler()
X_scaled=scalar.fit_transform(X)
```

This is how our data looks now after scaling. Great, now we will check for multicollinearity using VIF(Variance Inflation Factor)

```python
X_scaled
```

```python
# Finding variance inflation factor in each scaled column i.e. X_scaled.shape[1]
```

```python
vif= pd.DataFrame()
vif["vif"]=[variance_inflation_factor(X_scaled,i) for i in range(X_scaled.shape[1])]
vif["Features"]=X.columns

# Let's check the values
vif
```

All the VIF are less than 5 and are very low. That means no multicollinearity. Now, we can go ahead with fitting our data to the model. Before, that, let's split our data in test and training set.

```python
x_train,x_test,y_train,y_test=train_test_split(X_scaled,y, test_size= 0.25, random_state= 355)
```

```python
log_reg= LogisticRegression()
log_reg.fit(x_train,y_train)
```

Let's see how well our model performs on the test dataset.

```python
y_pred= log_reg.predict(x_test)
```

```python
accuracy=accuracy_score(y_test,y_pred)
accuracy
```

```python
# Confusion Matrix
conf_mat= confusion_matrix(y_test,y_pred)
conf_mat
```

```python
true_positive=conf_mat[0][0]
false_positive=conf_mat[0][1]
false_negative=conf_mat[1][0]
true_negative=conf_mat[1][1]
```

```python
# Breaking down the formula for accuracy
Accuracy=(true_positive+true_negative)/(true_positive+false_positive+false_negative+true_negative)
Accuracy
```

```python
# Precision
Precision=true_positive/(true_positive+false_positive)
Precision
```

```python
# Recall
Recall= true_positive/(true_positive+false_negative)
Recall
```

```python
# F1 Score
F1_Score= 2*(Recall*Precision)/(Recall+Precision)
F1_Score
```

```python
# Area Under Curve
auc= roc_auc_score(y_test, y_pred)
auc
```

# ROC

```python
fpr,tpr,thresholds=roc_curve(y_test,y_pred)
```
