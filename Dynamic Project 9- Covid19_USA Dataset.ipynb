From above, we can observe that null values are present in the dataset.
Also, all these null values are present in numeric columns. Hence, I will impute these null values with median.

```python
# Imputing the missing value with their mean.
df['Lat'].fillna(df['Lat'].median(),inplace=True)
df['Long_'].fillna(df['Long_'].median(),inplace=True)
df['Recovered'].fillna(df['Recovered'].median(),inplace=True)
df['Incident_Rate'].fillna(df['Incident_Rate'].median(),inplace=True)
df['People_Tested'].fillna(df['People_Tested'].median(),inplace=True)
df['People_Hospitalized'].fillna(df['People_Hospitalized'].median(),inplace=True)
df['Mortality_Rate'].fillna(df['Mortality_Rate'].median(),inplace=True)
df['Testing_Rate'].fillna(df['Testing_Rate'].median(),inplace=True)
df['Hospitalization_Rate'].fillna(df['Hospitalization_Rate'].median(),inplace=True)
```

```python
# Checking null values again via heatmap
sns.heatmap(df.isnull())
plt.show()
```

Observation: All the null values have been filled with median

```python
# Treating skewness via square root and cube root method.
df.skew()
for col in df.skew().index:
    if col in df.describe().columns:
        if df[col].skew()>0.55:
            df[col]=np.sqrt(df[col])
        if df[col].skew()<-0.55:
            df[col]=np.cbrt(df[col])
```

# Removing Outliers

```python
from scipy.stats import zscore
z_score=abs(zscore(df))
print(df.shape)
df_final=df.loc[(z_score<3).all(axis=1)]
print(df_final.shape)
# All outliers have been removed.
```

Note: Before outliers removal, we have 58 rows and 14 columns. After removal of outliers, we have 48 rows and 14 columns.

# Model Training

```python
df_x=df_final.drop(columns=["Deaths"])
y=pd.DataFrame(df_final["Deaths"])
print(df_x.shape)
print(y.shape)
```

```python
#scaling the input variable
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(df_x)
x=pd.DataFrame(x,columns=df_x.columns)
```

```python
# Importing our Models Library
from sklearn.linear_model import LinearRegression,ElasticNet
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor

# Importing Error Metrics
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
#Breaking our input and target variable into train and test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=42)
```

```python
print(x_train.shape, '\t',x_test.shape)
```

```python
print(y_train.shape,'\t',y_test.shape)
```

```python
# Using algorithms via for loop

model=[LinearRegression(),DecisionTreeRegressor(),KNeighborsRegressor(),SVR(),ElasticNet(),RandomForestRegressor(),AdaBoostRegressor(),GradientBoostingRegressor()]
for m in model:
    m.fit(x_train,y_train)
    print('Score of',m, 'is:',m.score(x_train,y_train))
    predm=m.predict(x_test)
    print('Error:')
    print('Mean absolute error:',mean_absolute_error(y_test,predm))
    print('Mean squared error:',mean_squared_error(y_test,predm))
    print('Root Mean Squared Error:',np.sqrt(mean_squared_error(y_test,predm)))
    print('r2_score:',r2_score(y_test,predm))
    print('**************************************************************************************************')
    print('\n')
```

```python
# Cross validating the models
from sklearn.model_selection import cross_val_score
model=[LinearRegression(),DecisionTreeRegressor(),KNeighborsRegressor(),SVR(),ElasticNet(),RandomForestRegressor(),AdaBoostRegressor(),GradientBoostingRegressor()]
for m in model:
    score=cross_val_score(m,x,y,cv=4,scoring='r2')
    print('Score of',m,'is:')
    print('Score:',score)
    print('Mean Score:',score.mean())
    print('Standard deviation:',score.std())
    print('*************************************************************************************************')
    print('\n')
```

# Linear Regression model is giving the best score.

```python
# Linear Regression model is giving best result so finding its best parameters using GridSearchCV
from sklearn.model_selection import GridSearchCV
lr=LinearRegression()
parameters={'copy_X':[bool,False],'normalize':[bool,False],'fit_intercept':[bool,True],'n_jobs':range(0,10)}
clf=GridSearchCV(lr,parameters)
clf.fit(x,y)
clf.best_params_
```

```python
# Using Linear Regression model with best its best parameters.

lr=LinearRegression(copy_X=bool,fit_intercept=bool,n_jobs=0,normalize=bool)
lr.fit(x_train,y_train)
print('Score:',lr.score(x_train,y_train))
predls=lr.predict(x_test)
print('\n')
print('Mean absolute error:',mean_absolute_error(y_test,predls))
print('Mean squared error:',mean_squared_error(y_test,predls))
print('Root Mean Squared Error:',np.sqrt(mean_squared_error(y_test,predls)))
print('\n')
print('r2_score:',r2_score(y_test,predls))
```

```python
# Cross validate Linear Regression.

score=cross_val_score(lr,x,y,cv=5,scoring='r2')
print('Score:',score)
print('Mean Score:',score.mean())
print('Standard deviation:',score.std())
```

Linear Regression is working well because it gives the best r2_score of 0.97, Mean absolute error of 2.27, Mean squared error of 9.14, Root Mean Squared Error of 3.02 and also it is also it is not suffering from overfitting or underfitting, as we have cross-validated the same. So we save Linear Regression as our final Model.

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression().fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

```python
#Saving the Linear Regression as our final model
from sklearn.externals import joblib
joblib.dump(lr,'covid19_USA_deaths.obj')
```

```python
# Loading the model
lr_from_joblib=joblib.load('covid19_USA_deaths.obj')

# using the loaded model to make predictions

lr_from_joblib.predict(x_test)
```
